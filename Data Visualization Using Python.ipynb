After downloading the data, we now need to import some libraries for data processing and visualization purpose. 

```python
# For thevisualization data processing is essential , let import a important data visualization library pandas
import pandas as pd
```

```python
#for statistical purpose and for better visualization lets import seaborn
import seaborn as sns
```

```python
#as seaborn creates different warning so to ignore the warning signals lets write few line 
import warnings
warnings.filterwarnings("ignore")
```

```python
#for different plots and charts visualization lets import pyplot from matplotlib 
import matplotlib.pyplot as plt 
```

```python
#Input the iris data set. The data set should be in the working directory or you should provide 
#the link for the data file
data = pd.read_csv("Iris.csv")
```

```python
#data is now imported, to view the first 5 data and last 5 data  
data.head()

```

```python
data.tail() #last 5 data
```

```python
#summary of the data 
data.describe()
#this will give you the min, max, standard deviation, quartiles and mean of the data as well as count of the data
```

```python
# Now we are interested to know how many species of we have in our data. 
data["Species"].value_counts()
#make sure you type the column name correctly. it is case sensetive
```

From this we know there are 50 counts for each species. Now lets break this into three gourps and see the properties of each groups 

```python
 
group = data.groupby('Species')
```

```python
group.describe()
```

Form this we can get the information for each parameter of each groups. 

Now lets plot the data into a scatter polt. 

```python
data.plot(kind = 'scatter', x = 'SepalLengthCm', y = "SepalWidthCm")
plt.show()
#this is generic plot, we can see the dots but actually couldnot visualize what is happening there. 
#we can add colurs for better visualization and see each species individually 
```

```python
group.plot(kind = "scatter", x = "SepalLengthCm", y = "SepalWidthCm", s= 25, color = "red")
plt.show()
```

```python
    #Now to join these 3 different plots with  seaborn library
    sns.jointplot(x="SepalLengthCm", y = "SepalWidthCm", data =data , size = 5)
    plt.show()
```

```python
    #Now to join these 3 different plots with different colors we use seaborn library
    sns.FacetGrid(data,hue = "Species", size=5)\
    .map(plt.scatter, "SepalLengthCm", "SepalWidthCm")\
    .add_legend()
    plt.show()
```

```python
# to have the better visualization we can change the color for each species 
#first assign the color to each species 
COL = {'color': ['blue', 'red', 'yellow']}
sns.FacetGrid(data, hue_kws =COL, hue = "Species", size = 5)\
.map(plt.scatter, "SepalLengthCm", "SepalWidthCm")\
.add_legend()
plt.show()
```

From this visualization we can clearly understand that the species Setosa have higher ratio of Sepal width to lenght as compare other species. Similarly we can visualiza the petal properties same way. 

```python
COL = {'color': ['blue', 'red', 'yellow']}
sns.FacetGrid(data, hue_kws =COL, hue = "Species", size = 5)\
.map(plt.scatter, "PetalLengthCm", "PetalWidthCm")\
.add_legend()
plt.show()
```

```python
COL = {'color': ['blue', 'red', 'yellow']}
sns.FacetGrid(data, hue_kws =COL, hue = "Species", size = 5)\
.map(plt.scatter, "SepalLengthCm", "PetalLengthCm")\
.add_legend()
plt.show()
```

```python
COL = {'color': ['blue', 'red', 'yellow']}
sns.FacetGrid(data, hue_kws =COL, hue = "Species", size = 5)\
.map(plt.scatter, "PetalWidthCm", "SepalWidthCm")\
.add_legend()
plt.show()
```

All these above visualization suggest that the species Setosa is quite easily differentiable among 3.

```python
#Now let see the data points with the box plots. 
sns.boxplot(x = "Species", y = "PetalLengthCm", data =data)
plt.show()
```

```python
#we can also add some data points on top of the box plot using Seaborn's extremly useful module called stripplot. 
#put jitter = true, to make sure the data points remains scattered rather than piled into a verticle line
sns.boxplot(x = "Species", y= "PetalLengthCm", data= data )
sns.stripplot(x= "Species", y= "PetalLengthCm", data= data, jitter = True, edgecolor= "gray")
plt.show()
```

```python
#Now lets add different colors to each box using the object artists

bx= sns.boxplot(x="Species", y="PetalLengthCm", data=data)
bx= sns.stripplot(x="Species", y="PetalLengthCm", data=data, jitter=True, edgecolor="gray")

boxtwo = bx.artists[2]
boxtwo.set_facecolor('red')
boxtwo.set_edgecolor('black')
boxthree=bx.artists[1]
boxthree.set_facecolor('yellow')
boxthree.set_edgecolor('black')

plt.show()
```

similar to scatter plot the violin plot also help in visualizing the density of the data. the denser reason are flatter and sparse regions are pointed 

```python
sns.violinplot(x="Species", y = "PetalLengthCm", data=data, size=7)
plt.show()
```

```python
#adding palette changes the color of each violin
sns.violinplot(x="Species", y = "PetalLengthCm", data=data, palette={"red", "yellow", "blue"}, size=7)
plt.show()
```

To analize the relationship between the species we can use pairplot. This will analyze each pair of species with each of their attribute

```python
sns.pairplot(data.drop("Id", axis =1), hue = "Species", size=4)
plt.show()
```

```python
sns.pairplot(data.drop("Id", axis =1), hue = "Species", size=4, diag_kind = "kde")
plt.show()
```
